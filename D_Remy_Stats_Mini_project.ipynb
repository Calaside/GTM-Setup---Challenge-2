```python
#Cleaning dataframes where its needed
#Grouping and merging for each analyze :

#1 product weight and size vs shipping time
#2 customer satisfaction vs shipping time
#3 sellers vs shipping time (related with the analyze above)
#4 products turnover and overall profitability
#5 sellers vs revenue/profitability
#6 geographic zones vs shipping time
```

```python
#1 order weight and volume vs shipping time  => merging orders-items data with reviews
df_analyze_product_weight = pd.merge(df_olist_orders_items, df_olist_products, on='product_id', how='inner')
df_analyze_product_weight_shipping_brut = pd.merge(df_analyze_product_weight, df_olist_orders, on = 'order_id', how='left')
df_analyze_product_weight_shipping = df_analyze_product_weight_shipping_brut.drop(columns = ['shipping_limit_date', 'order_item_id', 'order_approved_at', 'seller_id', 'price', 'product_category_name', 'customer_id'])
df_analyze_product_weight_shipping['product_weight_g'] = df_analyze_product_weight_shipping['product_weight_g'].astype(np.float64)
df_analyze_product_weight_shipping = df_analyze_product_weight_shipping.drop(columns=['product_length_cm', 'product_height_cm', 'product_width_cm'])
df_analyze_product_weight_shipping.head()
```

```python
df_order_weight_size = df_analyze_product_weight_shipping.groupby('order_id').agg(
    {'product_weight_g':'sum',
     'product_volume_cm3':'sum',
     'shipping_time':'max'})
df_order_weight_size
```

```python
#tester la correlation entre moyenne de delai et le volume et le poids d'une part
df_analyze_product_weight_shipping.info()
#End of 1rst dataset for this analyze
```

```python
#2 customer satisfaction vs shipping time
df_analyze_shipping_satisfaction = pd.merge(df_olist_orders, df_olist_orders_reviews, on='order_id', how='left')
df_analyze_shipping_satisfaction = df_analyze_shipping_satisfaction.drop(columns=['review_id','review_creation_date','review_answer_timestamp'])
df_analyze_shipping_satisfaction['shipping_time'] = df_analyze_shipping_satisfaction['order_delivered_customer_date'] - df_analyze_shipping_satisfaction['order_purchase_timestamp']
df_analyze_shipping_satisfaction.head()
```

```python
#tests a faire : tester la correlation entre le temps de livraison et la satisfaction client
df_analyze_shipping_satisfaction.info()
#End of 2nd dataset for this analyze
```

```python
#3 sellers vs shipping time (related with the analyze above)
```

```python
df_olist_orders_overall_trans = pd.merge(df_olist_orders, df_olist_orders_reviews, on='order_id', how='left')
df_olist_orders_overall = pd.merge(df_olist_orders_overall_trans, df_olist_orders_items, on= 'order_id', how = 'left')
df_olist_orders_overall = df_olist_orders_overall.groupby('seller_id').agg(
    {'order_id':'count',
     'shipping_time':'mean',
     'review_score':'mean',
     'freight_value':'mean'})
#
df_olist_orders_overall
```

```python
#adding some geographics data for some fancy map analysies
df_analyze_shipping_sellers = pd.merge(df_olist_orders_overall, df_olist_sellers, on='seller_id', how='left')
df_analyze_shipping_sellers
```

```python
#4 products turnover and overall profitability
```

```python
df_olist_categories_revenue = df_olist_orders_items.merge(df_olist_products, on='product_id', how='left')

df_olist_categories_revenue = df_olist_categories_revenue.groupby('product_category_name').agg({
     'order_id':'count',
     'freight_value':'sum' ,
     'price':'sum' ,
     'product_id':'count'})
df_olist_categories_revenue['margin'] = df_olist_categories_revenue['price'] - df_olist_categories_revenue['freight_value']
df_olist_categories_revenue['quantity'] = df_olist_categories_revenue['order_id']

df_olist_categories_revenue
#juste a ploter la, pas vraiment de test a faire mais simplement montrer la repartition et les best-sellers
```

```python
#5 geographic zones vs shipping time
```

```python
df_olist_items_sellers = pd.merge(df_olist_orders_items, df_olist_sellers, on='seller_id', how='left')
df_olist_items_sellers_orders = pd.merge(df_olist_items_sellers, df_olist_orders, on='order_id', how='left')
df_analyze_geographic = pd.merge(df_olist_items_sellers_orders, df_olist_customers, on='customer_id', how='left')
df_analyze_geographic.drop(['product_id',
                            'shipping_limit_date',
                            'price',
                            'freight_value',
                            'order_purchase_timestamp',
                            'order_approved_at',
                            'order_delivered_carrier_date',
                            'order_delivered_customer_date',
                            'order_estimated_delivery_date'],
                           axis=1,
                           inplace= True)

df_analyze_geographic_sellers = df_analyze_geographic.groupby('seller_id').agg({
    'order_id':'count',
    'shipping_time':'mean',
    'seller_zip_code_prefix':'first',
    'seller_city':'first',
    'customer_zip_code_prefix':'first',
    'customer_city':'first',
    'customer_state':'first'
})

df_analyze_geographic_customers = df_analyze_geographic.groupby('customer_id').agg({
    'order_id':'count',
    'shipping_time':'mean',
    'customer_zip_code_prefix':'first',
    'customer_city':'first',
    'customer_state':'first'
})
```

```python
#To test => we have a dependent and categorical variable : review score
#        => we have a dependent and continuous variable : shipping time
#so, to see if the differences between each mean shipping time for the different review score are significant
#I'll use the Anova test after checking if there is a normal value distribution
```

```python
from scipy import stats

test = stats.normaltest(df_analyze_test[df_analyze_test['review_score'] == 5]['shipping_time'], axis=0, nan_policy='omit')
print(test)
```

```python
from scipy.stats import f_oneway
stats.f_oneway(df_analyze_test['review_score'] == 1,
               df_analyze_test['review_score'] == 2,
               df_analyze_test['review_score'] == 3,
               df_analyze_test['review_score'] == 4,
               df_analyze_test['review_score'] == 5)
```
