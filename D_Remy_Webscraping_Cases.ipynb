```python
cases_data = {}
```

```python
cases = []
dates = []

case_counter = 0
date_counter = 0
params = {'page_no' : 1}
```

```python
case_counter = 0
url = 'https://www.gov.uk/employment-tribunal-decisions?'

response = requests.get(url, params=params)
soup = BeautifulSoup(response.text, 'html.parser')
while case_counter <= 1000:
  for case_alone in soup.find_all('a', {'class' : 'govuk-link govuk-link--no-underline'}):
      cases.append(case_alone.text.strip())
  for date_alone in soup.find_all('li', {'class' : 'gem-c-document-list__attribute'}):
    dates.append(date_alone.text.strip())
    cases_data[case_counter] = {
      'case' : cases[case_counter],
      'date' : dates[case_counter]}
  case_counter += 1
  params['page_no'] += 1
```
